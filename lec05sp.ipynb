## ¿Qué es?

* Los rangos son arreglos de números consecutivos.
### Funciones de numpy
- `np.arange(end):` Un arreglo de enteros incrementales desde 0 hasta `end`
- `np.arange(start, end):` Un arreglo de enteros incrementales desde `start` hasta `end`
- `np.arange(start, end, step):` Un rango con saltos entre valores consecutivos


El rango siempre incluye `start` pero no `end`

## Creando una tabla desde cero ##

```python
calles = make_array('Amigos', 'Embrujo', 'La 12', 'Calvario')
calles
```

```python
Table()
```

```python
candelaria = Table().with_column('Calles', calles)
candelaria
```

```python
candelaria.with_column('Cuadras desde la U', np.arange(4))
```

```python
precios= make_array(10000 , 8500 , 9800 , 12000)
color= make_array('azul' , 'rojo' , 'naranja' , 'verde')
candelaria.with_column('Cuadras desde la U', np.arange(4)).with_column('precios' , precios).with_column('color', color) #comillas caracteres, si comillas numeros
```

```python
candelaria
```

```python
candelaria = candelaria.with_column('Cuadras desde la U', np.arange(4))
candelaria
```

```python
Musica= make_array('rock' , 'vallenato' , 'ranchera' , 'regueton')
Seguridad= make_array('alta' , 'baja' , 'media' , 'alta')
TipoComida= make_array('mar' , 'vegetariano', 'asadero' , 'corriente')
Aforo= make_array(12,10,20,15)
candelaria.with_column('Cuadras desde la U', np.arange(4)).with_column('precios' , precios).with_column('Musica' , Musica).with_column( 'Seguridad' , Seguridad).with_column( 'TipoComida' , TipoComida).with_column( 'Aforo' , Aforo)
```

## Seleccionando data de una columna ##

```python
movies = Table.read_table('movies_by_year_with_ticket_price.csv')
movies.show()
```

```python
gross_in_dollars = movies.column('Total Gross') * 1e6
tix_sold = gross_in_dollars / movies.column('Average Ticket Price')
```

```python
movies = movies.with_column('Tickets sold', tix_sold)
```

```python
movies.show(4)
```

```python
movies.set_format('Tickets sold', NumberFormatter)
```

```python
movies.plot('Year', 'Tickets sold')
```

```python
movies.where('Year', are.between(2000, 2005))
```

```python
movies.where('Year', 2002)
```

```python
movies.where('Year', are.equal_to(2002))
```

```python
movies.where('#1 Movie', are.containing('Harry Potter'))
```

```python
movies.take(np.arange(2, 5))
```
